```python
#Function Which takes the arguments as the data, the columns for analysis and the directory to capture and store the graphs
def Graphs(Data, Variables=[], Directory=os.getcwd()):

    #Avoiding anyother data types as we want to plot only for numeric and categorical data
    dtype1 = ['int64', 'float64']
    dtype2 = ['O']

    #Checking if the Directory Exists
    if os.path.exists(Directory + 'Histogram') == False and os.path.exists(Directory +'Barplot') == False and os.path.exists(Directory +'Boxplot') == False:

        #Create the Directory Path if it does not Exists
        os.mkdir(Directory + 'Histogram')
        os.mkdir(Directory + 'Barplot')
        os.mkdir(Directory + 'Boxplot')

        #Change the Directory to Changed Directory
        os.chdir(Directory)

    #Checking if there are no columns sepcified
    if Variables == []:

        #Replacing it with Default variables i.e all the variables
        Variables = Data.columns

    #loop to run through all the variables of interest
    for i in Variables:

        #Checking for Numeric variables which cannot be categorical
        if Data[i].dtype in dtype1 and len(Data[i].unique()) > 15:

            #For Assigning title to a graph
            plt.title('Histogram of '+ i)

            #For Labeling the Axis
            plt.xlabel(i)

            #Plotting histogram for the numeric variable
            plt.hist(Data[i], color='green', edgeColor='black')

            #To save the graph as a picture of png formata in the given parameter i.e Directory
            plt.savefig(Directory + 'Histogram/' + 'Histogram of ' + i + '.png')

            plt.show()
            #Close the previous graph to view another graph
            #plt.close()

            #For Assigning title to a graph
            plt.title('Box Plot')

            #For Labeling the Axis
            plt.xlabel(i)

            #Plotting Box Plot for the numeric variable
            plt.boxplot(Data[i], patch_artist=True, vert=False)

            #To save the graph as a picture of png formata in the given parameter i.e Directory
            plt.savefig(Directory + 'Boxplot/' + 'Boxplot of ' + i + '.png')

            plt.show()
            #Close the previous graph to view another graph
            #plt.close()

        #Numeric data which are finite and can be considered as categorical variables
        elif Data[i].dtype in dtype1 and len(Data[i].unique()) < 15:

            #Plotting Bar Plot for such variables
            plt.bar(Data[i].unique(),
                    height=np.array(
                        pd.crosstab(Data[i], columns="Counts").Counts),
                    width=len(Data[i].unique()) * 0.1, color = 'red')

            #X axis will be the Number of Categories
            plt.xlabel(i)
            #To save the graph as a picture of png formata in the given parameter i.e Directory
            plt.savefig(Directory + 'Barplot/' + 'BarPlot of ' + i + '.png')

            plt.show()
            #Close the previous graph to view another graph
            #plt.close()

        #If it is a Categorical Variable
        elif Data[i].dtype in dtype2 and len(Data[i].unique()) < 1:

            #Plotting Bar Plot for the Categorical variable
            plt.bar(
                Data[i].unique(),
                height=np.array(pd.crosstab(Data[i], columns="Counts").Counts),
                width=len(Data[i].unique()) * 0.1, color = 'red'
            )

            #X axis will be the Number of Categories
            plt.xlabel(i)

            #To save the graph as a picture of png formata in the given parameter i.e Directory
            plt.savefig(Directory + 'Barplot/' + 'BarPlot of ' + i + '.png')

            plt.show()
            #Close the previous graph to view another graph
            #plt.close()
```
